```python
import pandas as pd
import re
import string
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#year vs number of reviews
df = pd.read_csv('extendedReviews.csv')
#convert reviewTime to year value
df['year'] = pd.DatetimeIndex(df['reviewTime']).year
#group by year
df = df.groupby(['year'])['overall'].sum()
df = df.reset_index()
df.columns = ['year','numOfReviews']
#plot
plt.title('Year vs Number of Reviews')
plt.plot(df["year"], df["numOfReviews"], color='cornflowerblue')
plt.xlabel("Year")
plt.ylabel("Review Count")
plt.legend()
```

```python
#number of reviews per star rating
df = pd.read_csv('extendedReviews.csv')
#group by star rating
df = df.groupby(['overall']).count()
df = df.reset_index()
#plot
plt.title('# of Reviews by Star Rating')
plt.pie(df["overall"], labels=df['overall'],autopct='%1.0f%%')
plt.legend()
```

```python
#Average star rating based on verification status
df = pd.read_csv('extendedReviews.csv')
#group by verification status and find mean star rating
df = df.groupby(['verified'])['overall'].mean()
df = df.reset_index()
#add labels to verification status
df['labels'] = ['Verified','Non-Verified']
#plot
plt.title('Verified vs Non-Verified Average Rating')
plt.bar(df["labels"], df["overall"], color='gold')
plt.ylabel("Star Rating")
plt.ylim(4,5)
plt.legend()
```

```python
#Average review length verse star rating

df = pd.read_csv('extendedReviews.csv')
#get length of each review
df['length'] = df['reviewText'].apply(len)
#get average review length for each star rating
df = df.groupby('overall').mean()
df = df.reset_index()

#plot
plt.title('Average Review Length vs Star Rating')
plt.bar(df["overall"], df["length"], color='forestgreen')
plt.ylabel("Review Length")
plt.xlabel("Star Rating")
plt.legend()
```
